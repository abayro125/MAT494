**3.5 K-Means**

K-means clustering is a common way of vector quantization that aims to partition n observations into k clusers where each observation belongs to a cluster. The nearest mean of the cluster acts as the prototype of the cluster. The within-cluster variances are minimized. 

Screen Shot 2022-10-24 at 11.21.22 AM.png

K-means clustering algorithm:
1) Clusters the data into k groups where k is predefined.
2) Select k points at random as cluster centers
3) Assign objects to their closest cluster according to the Euclidean distance function. 
4) Calculate the centroid or mean of all objects in each cluster
5) Repeat steps 2, 3 and 4 until the same points are assignment to each cluster



```python
from sklearn.datasets import make_blobs
import pandas as pd

dataset, classes = make_blobs(n_samples=200, n_features=2, centers=4, cluster_std=0.5, random_state=0)
# make as panda dataframe for easy understanding
df = pd.DataFrame(dataset, columns=['var1', 'var2'])

from sklearn.cluster import KMeans
kmeans = KMeans(n_clusters=4, init='k-means++', random_state=0).fit(df)

import seaborn as sns
import matplotlib.pyplot as plt
sns.scatterplot(data=df, x="var1", y="var2", hue=kmeans.labels_) #show scatter plot
plt.show()

sns.scatterplot(data=df, x="var1", y="var2", hue=kmeans.labels_)
plt.scatter(kmeans.cluster_centers_[:,0], kmeans.cluster_centers_[:,1], 
            marker="X", c="r", s=80, label="centroids") #plot the centers of the k-means clusters
plt.legend()
plt.show()
```
