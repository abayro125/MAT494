3.3.2 Convexity and Global Minimizers

A function is considered convex if the line ssegment between any two points on the graph of the function lies above he graph between the two points. 

Definition 3.3.13 (Convex Function) A function f : Rd R is convex if, for
all x,y  Rd and all   [0,1],
f ((1)x+y)  (1) f (x)+ f (y). 

```python
import numpy as np
import matplotlib.pyplot as plt #import matlib plot library

x = np.linspace(-1, 2) #define x axis

plt.figure(1, figsize=(3, 2.5))
plt.clf()

# A convex function
plt.plot(x, x**2, linewidth=2) #plot the convex function
plt.text(-.7, -.6**2, '$f$', size=20)

# The tangent in one point
plt.plot(x, 2*x - 1)
plt.plot(1, 1, 'k+')
plt.text(.3, -.75, "Tangent to $f$", size=15)
plt.text(1, 1 - .5, 'C', size=15)

# Convexity as barycenter
plt.plot([.35, 1.85], [.35**2, 1.85**2])
plt.plot([.35, 1.85], [.35**2, 1.85**2], 'k+')
plt.text(.35 - .2, .35**2 + .1, 'A', size=15)
plt.text(1.85 - .2, 1.85**2, 'B', size=15)

plt.ylim(ymin=-1)
plt.axis('off')
plt.tight_layout()

# Convexity as barycenter
plt.figure(2, figsize=(3, 2.5))
plt.clf()
plt.plot(x, x**2 + np.exp(-5*(x - .5)**2), linewidth=2)
plt.text(-.7, -.6**2, '$f$', size=20)

plt.ylim(ymin=-1)
plt.axis('off')
plt.tight_layout()
plt.show()
```

Lemma 3.3.14 (Affine Functions are Convex) Let w  Rd and b  R. The
function f (x) = wT x+b is convex.

Lemma 3.3.15 (First-Order Convexity Condition) Let f : Rd R be continuously
differentiable. Then f is convex if and only if for all x,y  Rd
f (y)  f (x)+f (x)T (yx).

Lemma 3.3.16 (Second-Order Convexity Condition) Let f : Rd R be twice
continuously differentiable. Then f is convex if and only if, for all x  Rd,
Hf (x) is PSD.

3.3.2.2 Global Minimizers of Convex Functions

For a convex function, sufficient (and therefore necessary) condition for minimizers
is f (x0) = 0.

Theorem 3.3.18 Let f : Rd R be a continuously differentiable, convex function.
If f (x0) = 0 then x0 is a global minimizer.

Theorem 3.3.19 (Global Minimizers of Convex Functions) Let f : Rd R be
a convex function. Then any local minimizer of f is also a global minimizer.

3.3.3 Gradient Descent

An iterative optimization algorithm for finding a local minimum of a differentiable function. 

```python
cur_x = 3 # The algorithm starts at x=3
rate = 0.01 # Learning rate
precision = 0.000001 #This tells us when to stop the algorithm
previous_step_size = 1 #
max_iters = 10000 # maximum number of iterations
iters = 0 #iteration counter
df = lambda x: 2*(x+5) #Gradient of our function


while previous_step_size > precision and iters < max_iters:
    prev_x = cur_x #Store current x value in prev_x
    cur_x = cur_x - rate * df(prev_x) #Grad descent
    previous_step_size = abs(cur_x - prev_x) #Change in x
    iters = iters+1 #iteration count
    print("Iteration",iters,"\nX value is",cur_x) #Print iterations
    
print("The local minimum occurs at", cur_x)
```
